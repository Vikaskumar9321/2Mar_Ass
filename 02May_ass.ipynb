A scatter plot is a type of data visualization that displays the relatioship between two numerical variables. It used dots or markers on two-dimensional plane, where each dot represents a single data point.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3+ np.random.normal(0,2,50)
y = 3+ np.random.normal(0,2,len(x))

plt.scatter(x, y)
plt.title("Scatter plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

#Ans.3

The subplot() function in matplotlib is used to create a grid of subplots within a single figure. it allows you to plot multiple plots in sparate sections of the figure, arranged in rows and columns.

```python
x1 = np.array([0,1,2,3,4,5])
y1 = np.array([0,100,200,300,400,500])

x2 = np.array([0,1,2,3,4,5])
y2 = np.array([50,20,40,20,60,70])

x3 = np.array([0,1,2,3,4,5])
y3 = np.array([10,20,30,40,50,60])

x4 = np.array([0,1,2,3,4,5])
y4 = np.array([200,350,250,550,450,150])

fig, axs = plt.subplots(2,2)
axs[0,0].plot(x1,y1)
axs[0,0].set_title("Line 1")
axs[0,1].plot(x1,y1)
axs[0,1].set_title("Line 1")
axs[1,0].plot(x1,y1)
axs[1,0].set_title("Line 1")
axs[1,1].plot(x4,y4)
axs[1,1].set_title("Line 4")

plt.tight_layout()
plt.show
```

A bar plot, also known as a bar chart, is a graphical reprentation of categorical data uing rectangular bars. It displays the relationship between categories or groups by showing the length or height of the bars corresponding to the values they represent. The length of each bar is proportional to the value it represents, allowing for easy comparison of different categories.

Bar plots are commonly used to :

1.) Compare the values of the different categories or groups.


2.) Show the distribution or frequency pf categorical data.


3.) Dispaly trends over time or across different conditions.


```python
company = np.array(["Apple", "Microsoft", " google","AMD"])
profit = np.array([3000,8000,1000,10000])

plt.bar(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit ($)")
plt.title("Profit by Company")
plt.show()
```

A box plot, also known as a box-and-whisker plot, is a graphical reprentation of a dataset that provides a summary of its distribution. It displays several statistical measures such as the median, quartiles, and outliers, allowing for a visual comparison of multiple datasets or identifying patterns within a single dataset.


A box plot consists of a rectangular box that spans the interquartile range(IQR), which represents the middle 50% of the data. Inside the box, a line is drawn at the median. The whiskers extend from the box to represent the range of the data, excluding Outliers are individual data points that fall significantly outside the expected range of values.

Box plots are useful for several reasons:

1.) They provide a concise summary of the data distribution, making it eassy to identify skewness, symmetry, and the presence of outliers.

2.) They allow for the comparison of multiple datasets on the same scale, enabling insights into differences in central tendency and variability.

3.)They help in identifying potential anomalies or unusual patterns within a dataset.  



```python
box1 = np.random.normal(100,10,200)
box2 = np.random.normal(90,20,200)

data = [box1, box2]
labels = ['Box1', 'Box2']

plt.boxplot(data, labels = labels)
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box plot')
plt.show()
```
